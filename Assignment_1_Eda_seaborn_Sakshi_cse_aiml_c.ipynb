```python
sns.kdeplot(x="bill_length_mm",y="flipper_length_mm",color="violet",fill=True,data=data)
```

```python
def kdeplot(data, x, ax):
    sns.kdeplot(data[x], ax=ax)

fig, ax = plt.subplots(1, 3, figsize=(27, 7))
for idx, y in enumerate(['body_mass_g', 'flipper_length_mm', 'bill_length_mm']):
    kdeplot(data, y, ax[idx])
```

POint Plot

```python
sns.lmplot(x='body_mass_g', y='flipper_length_mm', hue='sex', data=data, palette='Set1')
plt.show()
```

```python
sns.set({'figure.figsize': (15, 15)})
sns.jointplot(x="bill_length_mm", y="flipper_length_mm", data=data, kind='kde', space=1,color="y",fill=True)
plt.show()
```


**VIOLIN PLOT**

```python

ndata=sns.load_dataset("tips")
ndata
```

```python
plt.figure(figsize=(10,7))
sns.violinplot(x="day",y="total_bill",data=ndata,hue="day",linewidth=3)
plt.show()
```

Here above plot the maximum total_bill on saturday

```python
plt.figure(figsize=(10,7))
sns.violinplot(x="time",y="total_bill",data=ndata,order=["Dinner","Lunch"],linewidth=3,inner="stick",color='r')
plt.show()
```

cat plot


```python
plt.figure(figsize=(10,7))
sns.catplot(x="time",y="total_bill",data=ndata, kind="swarm",color='r', hue='sex')
plt.show()
```
